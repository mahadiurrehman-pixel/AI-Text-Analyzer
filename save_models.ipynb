```python
import torch 
from transformers import pipeline
from transformers import AutoTokenizer , AutoModelForSequenceClassification
```

```python
sentiment_classifier=pipeline('sentiment-analysis')
```

```python
ner = pipeline('ner',model='dslim/bert-base-NER')
zero_shot_classifier=pipeline("zero-shot-classification",model="facebook/bart-large-mnli")
```

```python
tokenizer_model =AutoTokenizer.from_pretrained('distilbert-base-uncased-finetuned-sst-2-english')
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
# Different directories
sentiment_dir = "Saved_Model/sentiment"
ner_dir = "Saved_Model/ner"
zero_shot_dir = "Saved_Model/zero_shot"
base_model_dir = "Saved_Model/base_model"

sentiment_classifier.save_pretrained(sentiment_dir)
ner.save_pretrained(ner_dir)
zero_shot_classifier.save_pretrained(zero_shot_dir)
tokenizer_model.save_pretrained(base_model_dir)
model.save_pretrained(base_model_dir)
```
